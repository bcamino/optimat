# Optimat Function Tests

Starter notebook for testing functions as they are developed.
This first version loads and prints the parsed YAML config object.

```python
from pathlib import Path
import sys

# Make `src/` importable whether the notebook is launched from the repo root or notebooks/.
cwd = Path.cwd().resolve()
candidates = [cwd, cwd.parent]
repo_root = next((p for p in candidates if (p / 'src').exists()), None)
if repo_root is None:
    raise RuntimeError('Could not locate repo root containing src/')

src_path = str(repo_root / 'src')
if src_path not in sys.path:
    sys.path.insert(0, src_path)

repo_root
```

```python
from optimat.config import read_yaml

config, base_dir = read_yaml(repo_root / 'dev_data' / 'input_test.yml', return_base_dir=True)
print(config)
config
```

## Compile and Inspect the Energy Model

Compile Buckingham + Ewald terms from the YAML config and inspect the compiled pair table.

```python
from optimat.energy import compile_energy_model

terms = compile_energy_model(config, base_dir=base_dir)

print('E0:', terms.E0)
print('onsite terms:', len(terms.onsite))
print('pair terms:', len(terms.pair))
print('meta:', terms.meta)

for key, value in list(sorted(terms.pair.items()))[:10]:
    print(key, value)

terms
```

## Manual Assignment Energy

Choose species on the variable sites manually and evaluate the compiled energy.

```python
manual_assignment = {
    0: 'Mg',
    1: 'Ca',
    2: 'Mg',
    3: 'Ca',
    4: 'O',
    5: 'O',
    6: 'O',
    7: 'O',
}

energy = terms.energy_of_assignment(manual_assignment)
print('Manual assignment:', manual_assignment)
print('Compiled energy:', energy)
```

## Build Canonical Problem

Build the solver-agnostic canonical problem object (structure, occupancy constraints, compiled energy terms, pair graph).

```python
from optimat.problem import build_problem

problem = build_problem(config, base_dir=base_dir)

print('name:', problem.name)
print('n_sites:', problem.n_sites)
print('variable_sites:', problem.variable_sites)
print('fixed_assignments:', problem.fixed_assignments)
print('site_groups:', problem.site_groups)
print('composition_constraints:', problem.composition_constraints)
print('pair_graph nodes:', len(problem.pair_graph))
print('pair_graph edges:', sum(len(v) for v in problem.pair_graph.values()) // 2)
print('energy meta:', problem.energy_terms.meta)

problem
```

## Exact Backend (Exhaustive)

Compile the canonical problem for the exact backend and enumerate feasible assignments to find the best energy.

```python
from optimat.backends.exact import compile_to_exact, run_exact

exact_compiled = compile_to_exact(problem)
exact_result = run_exact(exact_compiled, top_k=5)

print('best_energy:', exact_result.best_energy)
print('best_assignment:', exact_result.best_assignment)
print('stats:', exact_result.stats)
print('top_k:', exact_result.top_k)

exact_result
```
